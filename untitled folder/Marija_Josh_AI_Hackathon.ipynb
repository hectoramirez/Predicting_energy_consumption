# Define  metrics <a id='metrics'></a>

To quantify our results lets define the loss function. The suggested ones is mea, mse, both of them are defined in sklearn


$ \rm{MAE}  = \frac{1}{n}\sum_i (\mid true_i  - predicted_i \mid	) $

$ \rm{MSE}  = \frac{1}{n}\sum_i \sqrt{(true_i  - predicted_i )^2} $



```python
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import mean_absolute_error as MAE
```

The relative error is then

```python
def relative_MAE(true, predicted):
    return np.abs(true - predicted)/true
```

```python
train = train_df['Demanda (W)'].values
valid = valid_df['Demanda (W)'].values[:240] #this is slow, thats why i limited it to first 240
import statsmodels.api as sm
history = [x for x in train]
predictions = list()
for t in range(len(valid)):
    model = sm.tsa.statespace.SARIMAX(history, order=(2, 0, 0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = valid[t]
    history.append(obs)
```

```python
np.mean(relative_MAE(valid, np.array(predictions)))
```

```python

plt.figure(figsize=(20, 10))
plt.plot(valid, label = 'true')
plt.plot(predictions, color='red', label='predicted')
plt.legend()
```

## TREE MODELS <a id='XGB'></a>
One can use any supervised learning model to deal with time series. For that we need our input and prediction to be

$ \rm{X} = [{x_1}^{t-n},{x_2}^{t-n}.. {y}^{t-n} ... {x_1}^{t-1},{x_2}^{t-1}.. {y}^{t-1}]$, and

$ \rm{Y} = {y}^t$


We will use XGBoost library for gradient boosted decision trees. It is not part of scikit-learn so should be installed via pip.

```python
! pip install XGBoost
```

```python
# From: https://www.tensorflow.org/tutorials/structured_data/time_series
def create_history_data(dataset, columns, target, history_steps = 3, future_steps = 1):
    data = []
    labels = []

    start_index = history_steps
    end_index = len(dataset) - future_steps

    for i in range(start_index, end_index):
        indices = range(i-history_steps, i)
        data.append(dataset[columns].iloc[indices].values)
        labels.extend(dataset[target].iloc[i:i+future_steps].values)

    return np.array(data), np.array(labels)
```

```python
columns = ['Temperature', 'Dew Point', 'Humidity', 'Wind', 'Speed', 'Gust',
       'Pressure', 'Precip. Rate.', 'Precip. Accum.', 'PV Tejado (W)',
       'PV2 Parque (W)', 'Demanda (W)']
history_steps = 3
future_steps  = 1
target        = 'Demanda (W)'

x_train, y_train = create_history_data(train_df, columns, target, history_steps, future_steps) 
x_valid, y_valid = create_history_data(valid_df, columns, target, history_steps, future_steps) 
```

```python
x_train.shape, y_train.shape
```

In order to use trees we have to reshape our x data into n_features x n_samples

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1] * x_train.shape[2])
x_valid = x_valid.reshape(x_valid.shape[0], x_valid.shape[1] * x_valid.shape[2])
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(
    max_depth=20,
    n_estimators=100,
    min_child_weight=300, 
    colsample_bytree=0.8, 
    subsample=0.8, 
    eta=0.3,    
    seed=42)
model.fit(
    x_train, 
    y_train, 
    eval_metric="mae", 
    eval_set=[(x_train, y_train), (x_valid, y_valid)], 
    verbose=False, 
    early_stopping_rounds = 10)
```

# RNN (LSTM)  MODEL <a id='LSTM'></a>
A recurrent normal network (RNN) is constructed to process a sequence of data in a series of steps, each of which takes as input a new data point as well as the output of the previous step. A long short-term memory network (LSTM) is a specific type of RNN designed to "remember" information over a long range of steps. The technical aspects of LSTMs are discussed in more detail here: https://colah.github.io/posts/2015-08-Understanding-LSTMs/.

To work with NNs we have to scale the data. The popular choices are either StandardScaling (mean 0 and variance 1) or MinMax scaling to some range, both are available in scikit_learn

```python

import tensorflow as tf
import keras
from keras.callbacks import EarlyStopping
from keras.regularizers import l2
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
columns = ['Temperature', 'Dew Point', 'Humidity', 'Wind', 'Speed', 'Gust',
       'Pressure', 'Precip. Rate.', 'Precip. Accum.', 'PV Tejado (W)',
       'PV2 Parque (W)', 'Demanda (W)']
history_steps = 24
future_steps  = 1
target        = ['Demanda (W)']

```

```python
normalize = MinMaxScaler().fit(train_df[columns])
norm_train_df = pd.DataFrame(columns=columns, data = normalize.transform(train_df[columns]))
norm_valid_df = pd.DataFrame(columns=columns, data = normalize.transform(valid_df[columns]))
targets_numpy_aux = train_df[target]
#if its is  1d we have to expand the last dimension
try:
  normalize_label = MinMaxScaler().fit(train_df[target].values)
except ValueError:
  normalize_label = MinMaxScaler().fit(train_df[target].values[:, None])
```

```python
x_train, y_train = create_history_data(norm_train_df, columns, target, history_steps, future_steps) 
x_valid, y_valid = create_history_data(norm_valid_df, columns, target, history_steps, future_steps)
#to be able to train the y has to be 2d, so we add one dimension in case it is not
if len(y_train.shape)==1:
  y_train = y_train[:, None]
  y_valid = y_valid[:, None]
```

```python
x_train.shape, y_train.shape
```

Construct a simple model 

```python
batch_size = 32
model = keras.models.Sequential()
model.add(keras.layers.LSTM(128, input_shape=(x_train.shape[1], x_train.shape[2]),activation='relu', return_sequences=False, kernel_regularizer=l2(1e-6), recurrent_regularizer=l2(1e-6), bias_regularizer=l2(1e-6)))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(len(target),  kernel_regularizer=l2(1e-6), bias_regularizer=l2(1e-6)))
opt = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)
model.compile(loss='mae', optimizer=opt)
```

```python
model.summary()
```

```python
epochs=100
Callbacks = [EarlyStopping(monitor='val_loss', min_delta=0, patience=10, restore_best_weights=True)]
model.fit(x_train, y_train, epochs=epochs, validation_data=(x_valid, y_valid), batch_size=batch_size, shuffle=False, callbacks = Callbacks)
```

```python
valid_predicted = model.predict(x_valid)
```

```python
y_true = normalize_label.inverse_transform(y_valid)
y_pred = normalize_label.inverse_transform(valid_predicted)
```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(valid_df.Date.iloc[history_steps+1:], y_true[:, 0], label='true')
plt.plot(valid_df.Date.iloc[history_steps+1:], y_pred[:, 0], label='predicted')
plt.legend()
```
