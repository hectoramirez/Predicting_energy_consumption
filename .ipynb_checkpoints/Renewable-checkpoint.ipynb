**Load csv and excel files using pandas**


> Note that I have parsed the day and time into one variable



```
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 
PATH = "gdrive/My Drive/Path_to_File/"

data = pd.read_csv(PATH+"Meteo.csv",parse_dates=[['Date', 'Time']] )
data.head()
```

```
EXCEL = PATH+"Datos_solar_y_demanda_residencial.xlsx"
df = pd.read_excel (EXCEL)

df.head()
```

```
df.describe()
```

---

**Get to a common timing for both datasets**



```
import datetime
def round_to_5min(t):
    delta = datetime.timedelta(minutes=t.minute%5, 
                               seconds=t.second, 
                               microseconds=t.microsecond)
    t -= delta
    if delta > datetime.timedelta(0):
        t += datetime.timedelta(minutes=5)
    return t
```

```
data['DateRound'] = data["Date_Time"].dt.round("5min")
df['DateRound'] = df['Date'].dt.round("5min")

data['Hour'] = data['DateRound'].dt.hour
data['Day'] = data['DateRound'].dt.dayofyear
data.head()
```

```
df.head()
```

```
data.columns
```

---

**Clean Data Frame from characters and columns that do not give information in this case: UV, Solar and Wind**

```
def clean(x):
    try:
        return  x.str.replace(r"[a-zA-Z\%\/²]",'')
    except:
        return x

data = data.apply(lambda x : clean(x))
data = data.drop(columns=['UV', 'Solar','Wind'])

data.head()
```

Checking variables :


```
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(18,4))

demanda_val = df['Demanda (W)'].values
time_val = df['DateRound'].values

temperatura_w = data['Temperature'].values
time_val_w = data['DateRound'].values

plt.plot(time_val,demanda_val)
plt.plot(time_val_w,temperatura_w)

sns.distplot(demanda_val, ax=ax[0], color='r')
ax[0].set_title('Just one random variable', fontsize=14)
ax[0].set_xlim([min(demanda_val), max(demanda_val)])



plt.show()
```

---


*Select* relevant variables for the final data frame :

* For meteo variables **all** in this example

* From energy data frame selection only 'Demanda (W)'

```
data_filtered = data[['DateRound', 'Temperature', 'Dew Point','Humidity', 'Humidity', 'Speed', 'Gust', 'Pressure', 'Precip. Rate.', 'Precip. Accum.','Hour', 'Day']]
df_filtered = df[['DateRound','Demanda (W)']]
```

Merge both pandas based on datetime



```
pfinal = pd.merge(df_filtered,data_filtered, on='DateRound')
pfinal.columns
# remove date from list
del pfinal['DateRound']
#show final result
pfinal.head()
```

---

**Make a nice plot with all the values for all variables**

```
from matplotlib import pyplot

totalVariables = len(pfinal.columns)
values = pfinal.values

# specify columns to plot
i = 1
# plot each column
pyplot.rcParams['figure.figsize'] = [20, 10]

pyplot.figure()
for group in range(0,totalVariables):
	pyplot.subplot(totalVariables, 1, i)
	pyplot.plot(values[:, group])
	pyplot.title(pfinal.columns[group], y=0.5, loc='right')
	i += 1
pyplot.show()
```
